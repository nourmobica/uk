```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url="https://quotes.toscrape.com/"
```

```python
soup=BeautifulSoup(requests.get(url).content,"html.parser")
soup.prettify
```

```python
a=[]
for quotes in soup.find_all("div",class_="quote"):
  text=quotes.find("span",class_="text").text
  by=quotes.find("span").text
  tags=quotes.find("div",class_="tags").text
  a.append([text,by,tags])
```

```python
a
```

```python
df=pd.DataFrame(a,columns=("text","by","tags"))
```

```python
df
```

```python
df.to_csv("kjad")
```
